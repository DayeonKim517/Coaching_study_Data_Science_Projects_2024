Í∞≠ÎßàÏù∏Îçî ÏÇ¨Ïù¥Ìä∏ÏóêÏÑúÎäî Ïó∞ÎèÑÎ≥Ñ, Íµ≠Í∞ÄÎ≥Ñ GDPÏôÄ Í∏∞ÎåÄÏàòÎ™Ö Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ≥µÌïòÍ≥† ÏûàÎäîÎç∞, ÎåÄÌëúÏ†ÅÏúºÎ°ú ÌååÏù¥Ïç¨ ÎùºÏù¥Î∏åÎü¨Î¶¨ Ï§ë 'seaborn'ÏóêÏÑú Ï†úÍ≥µÎêòÎäî ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûàÏäµÎãàÎã§. Ïò§ÎäòÏùÄ Ïù¥ Îç∞Ïù¥ÌÑ∞Î•º ÌôúÏö©Ìï¥ Î¨∏Ï†úÎ•º ÌíÄÏñ¥Î≥¥Í≤†ÏäµÎãàÎã§.  
Îã§ÏùåÏùò Îç∞Ïù¥ÌÑ∞Îäî Ïó∞ÎèÑ, Íµ≠Í∞ÄÎ≥Ñ Í∏∞ÎåÄÏàòÎ™ÖÏùÑ ÎÇòÌÉÄÎÇ¥Í≥† ÏûàÎäî Îç∞Ïù¥ÌÑ∞ ÏûÖÎãàÎã§. 2011ÎÖÑ Î∂ÄÌÑ∞Ïùò Ïó∞ÎèÑÎ≥Ñ, Íµ≠Í∞ÄÎ≥Ñ ÌèâÍ∑† Í∏∞ÎåÄÏàòÎ™ÖÏùÑ Íµ¨Ìï¥Ï£ºÏÑ∏Ïöî. (2011ÎÖÑÎèÑ Ìè¨Ìï®ÎêòÍ≤å Íµ¨Ìï©ÎãàÎã§.)

* [ÌïÑÏàò Ï°∞Í±¥] groupby ÎÇò pivot_tableÏùÑ ÌôúÏö©Ìï©ÎãàÎã§. groupby Î°ú Íµ¨Ìï† Îïå unstack()Ïù¥ÎùºÎäî Í∏∞Îä•ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏïÑÎûòÏôÄ Í∞ôÏù¥ Ïª¨ÎüºÏóê Ïù∏Îç±Ïä§ Í∞íÏùÑ Ïò¨Î†§ÏÑú ÌëúÍ∏∞Ìï† Ïàò ÏûàÏäµÎãàÎã§.
* ÎßåÏïΩ ÌñâÏóêÎäî 'Ïó∞ÎèÑ'Í∞Ä, Ïó¥Ïóê 'Íµ≠Í∞Ä'Í∞Ä Îì§Ïñ¥ÏûàÍ≥† ÏàòÏπò Îç∞Ïù¥ÌÑ∞Ïùò Í≤∞Í≥ºÍ∞íÏù¥ ÏïÑÎûòÏôÄ Í∞ôÎã§Î©¥, Ï∂úÎ†•ÌòïÌÉúÎäî Ï°∞Í∏à Îã¨ÎùºÎèÑ Í¥úÏ∞ÆÏäµÎãàÎã§.

````python
import pandas as pd
````

````python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df.head()
````

````python
df.info()
````

````python
df_y = df[df['Year'] >= 2011] # 2011ÎÖÑ Ïù¥ÏÉÅÏù∏ ÌñâÎßå ÎÇ®Í∏∞Í∏∞
df_y.head()
````

## pivot_table

````python
# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ df_yÎ•º pivot_tableÎ°ú Î≥ÄÌôò
# indexÎäî 'Year' Í∏∞Ï§ÄÏúºÎ°ú ÏÑ§Ï†ï
# columnsÎäî 'Country' Í∏∞Ï§ÄÏúºÎ°ú ÏÑ§Ï†ï
# valuesÎäî 'Life_Expectancy' Í∞íÏùÑ ÏÇ¨Ïö©

pivot_df = df_y.pivot_table(index='Year', columns='Country', values='Life_Expectancy')
pivot_df
````

## groupby

````python
# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ df_yÎ•º 'Year'ÏôÄ 'Country' Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôî
# Í∑∏Î£πÌôîÎêú Í∞Å Í∑∏Î£πÏóêÏÑú 'Life_Expectancy' Í∞íÏùò ÌèâÍ∑†ÏùÑ Í≥ÑÏÇ∞
# Í≥ÑÏÇ∞Îêú Í≤∞Í≥ºÎ•º unstack() Î©îÏÜåÎìúÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÌòïÌÉú Î≥ÄÌôò

group_df = df_y.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack()
group_df
````
